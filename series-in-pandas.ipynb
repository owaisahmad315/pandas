```python
import pandas as pd
```

```python
songs2 = pd.Series([145, 142,38,13], name = 'counts')
songs2
```

```python
songs2 = pd.Series([145, 142,38,13], name = 'counts', index = ['paul', 'jhon', 'owais', 'someone'])
songs2
```

```python
# The NaN value
nan_sar = pd.Series([2, None],
                    index = ['oho', 'Clapton'])
nan_sar
```

```python
import numpy as np
numpy_ser = np.array([145,142,38,13])
songs2[1]
numpy_ser[1]
```

```python
numpy_ser.mean()
```

```python
songs2.mean()
```

```python
mask = songs2 > songs2.median()
mask
```
